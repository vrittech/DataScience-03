```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier  # KNN Classifier
from sklearn.naive_bayes import GaussianNB
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
from sklearn.datasets import make_moons, make_circles
```

```python
# X, y = make_moons(n_samples=100, noise=0.15, random_state=666)
X, y = make_circles(n_samples=100, noise=0.15, random_state=666, factor=0.2)
```

```python
X[:5]
```

```python
y[:5]
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1], c = 'r')
plt.scatter(X[y==1, 0], X[y==1, 1], c = 'b')
plt.show()
```

```python
log_reg = LogisticRegression(random_state=666, n_jobs=-1)
```

```python
log_reg.fit(X, y)
```

```python
plot_decision_regions(X, y, log_reg)
plt.show()
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(X, y)

plot_decision_regions(X, y, dtree)
plt.show()
```

```python
knn = KNeighborsClassifier()
knn.fit(X, y)
 
plot_decision_regions(X, y , knn)
plt.show()
```

```python
nav = GaussianNB()
nav.fit(X, y)
 
plot_decision_regions(X, y , nav)
plt.show()
```
