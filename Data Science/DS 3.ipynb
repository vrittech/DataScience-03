```python
meu = np.array([3, 2]) # mean_x, mean_y
sigma = np.array([
    [10, 0], # sigma_x
    [0, 2] # sigma_y
])
```

# Categorical Data

```python
labels = np.array(['Nepal', 'India', 'China', 'Nepal', 'China',
                  'Nepal', 'Nepal', 'India', 'China'])
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
label_encoder = LabelEncoder()
ordinal_encoder = OrdinalEncoder()
one_hot_encoder = OneHotEncoder(sparse_output=False)
```

```python
label_encoder.fit(labels)
```

```python
countries = label_encoder.transform(labels)
countries
```

```python
label_encoder.classes_
```

```python
labels.reshape(-1, 1)
```

```python
one_hot_encoder.fit_transform(labels.reshape(-1, 1)) # (row_operarion, column_coperation)
```

```python
one_hot_encoder.categories_
```

```python
ordinal_encoder.fit_transform(labels.reshape(-1, 1))
```

```python
ordinal_encoder.categories_
```
