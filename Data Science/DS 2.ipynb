```python
meu = 6 # mean
sigma = 1 # Standard Deviation

normal = np.random.normal(meu, sigma, 1000)
normal
```

```python
# Technique for data scaling
class StandardScaler:
    def __init__(self, ):
        self.mean = None
        self.sigma = None
    def fit(self, data):
        self.mean = data.mean()
        self.sigma = data.std()
    def transform(self, data):
        return (data - self.mean) / self.sigma
    def fit_transform(self, data):
        self.fit(data)
        transformed = self.transform(data)
        return transformed
```

```python
normal
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(normal)
```

```python
scaled_norm = scaler.transform(normal)
```
