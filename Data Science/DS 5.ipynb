```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.linspace(10, 100, 15)
```

```python
y1 = np.linspace(10, 100, 15)
y2 = y1 ** 1.25
y3 = y1 ** 1.56

x = np.linspace(1, 15, 15)
```

```python
x
```

```python
plt.scatter(x, y1)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.title('My Plot')
plt.xlabel('Experience (in years)')
plt.ylabel('Yearly Income')
plt.plot(x, y1, color='r', label='Hari Bahadur', marker='X')
plt.plot(x, y2, color='b', label='Mai Saap', marker='o')
plt.plot(x, y3, color='g', label='Kaji', marker='+')
plt.legend()
plt.show()
```

```python
yatayat = ['Sajha Bus', 'Nepal Yatayat', 'Chakrapath Parikrama', 'Majur Yatayat', 'Agni Yatayat']
buses = [50, 100, 120, 80, 90]
```

```python
custom_color = ['green', 'brown', 'hotpink', 'r', '#ff00da']
```

```python
plt.bar(yatayat, buses, color=custom_color)
plt.show()
```

```python
plt.barh(yatayat, buses, color=custom_color)
plt.show()
```

```python
yatayat
```

```python
explode_val = [0, 0, 0.1, 0, 0.05]
plt.pie(x=buses, labels=yatayat,colors=custom_color, autopct='%.1f%%', explode=explode_val)
plt.savefig('Pie_Chart.png')
plt.show()
```
