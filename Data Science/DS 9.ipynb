```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

* Mean Absolute Error(MAE): $\frac{\sum_{i=1}^{N}|y_{actual} - y_{pred}|}{N}$
* Mean Squared Error(MSE): $\frac{\sum_{i=1}^{N}(y_{actual} - y_{pred})^2}{N}$
* R2: 1 - $\frac{\sum_{i=1}^{N}(y_{actual} - y_{pred})^2}{\sum_{i=1}^{N}(y_{actual} - mean)^2}$

```python
X = np.random.normal(10, 2, 120) # mean, std, size
y = X + np.random.normal(3, 0.5, 120)
```

```python
X[:5]
```

```python
y[:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
plt.scatter(X_train, y_train, c='b')
plt.scatter(X_test, y_test, c='r')
plt.show()
```

```python
m, c, r2, _, err = linregress(X_train, y_train)
```

```python
m
```

```python
c
```

```python
r2
```

```python
reg_line_train = m * X_train + c
```

```python
plt.scatter(X_train, y_train, c='b')
plt.plot(X_train, reg_line_train, c='r')
plt.show()
```

```python
reg_line_test = m * X_test + c
```

```python
plt.scatter(X_test, y_test, c='b')
plt.plot(X_test, reg_line_test, c='r')
plt.show()
```

```python
def predict(X_new):
    return m * X_new + c
```

```python
predict(8)
```

```python
predict(14)
```

```python
predict(-3)
```

```python
linear_reg = LinearRegression(n_jobs=-1, )
```

```python
X_train
```

```python
linear_reg.fit(X_train.reshape(-1, 1), y_train)
```

```python
linear_reg.coef_
```

```python
linear_reg.intercept_
```

```python
y_pred_test = linear_reg.predict(X_test.reshape(-1, 1),)
y_pred_train = linear_reg.predict(X_train.reshape(-1, 1),)
```

```python
mean_absolute_error(y_test, y_pred_test)
```

```python
mean_absolute_error(y_train, y_pred_train)
```

```python
r2_score(y_test, y_pred_test)
```

```python
r2_score(y_train, y_pred_train)
```
